```python
#Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing dataset from csv to data frame
visual_data = pd.read_csv('machine_learning_set.csv')
```

```python
visual_data
```

```python
#Exploring daily_non_vehicle_traffic on holidays
holidays = visual_data.loc[visual_data.holiday != 'none']
plt.figure(figsize=(10,8))
sns.boxplot(y='holiday',x='daily_non_vehicle_traffic', data = holidays)
plt.show()
```

```python
#Exploring daily_non_vehicle_traffic on holidays
holidays = visual_data.loc[visual_data.holiday != 'none']
plt.figure(figsize=(10,8))
sns.barplot(y='holiday',x='daily_non_vehicle_traffic', data = holidays)
plt.show()
```

```python
#Exploring total_vehicle_volume_daily on holidays
holidays = visual_data.loc[visual_data.holiday != 'none']
plt.figure(figsize=(10,8))
sns.boxplot(y='holiday',x='total_vehicle_volume_daily', data = holidays)
plt.show()
```

```python
#Exploring total_vehicle_volume_daily on holidays
holidays = visual_data.loc[visual_data.holiday != 'none']
plt.figure(figsize=(10,8))
sns.barplot(y='holiday',x='total_vehicle_volume_daily', data = holidays)
plt.show()
```

```python
#Plotting total_vehicle_volume_daily over month
plt.figure(figsize=(18,12))
sns.barplot(x='month', y = 'total_vehicle_volume_daily', data = visual_data)
plt.show()
```

```python
#Plotting daily_non_vehicle_traffic over month
plt.figure(figsize=(18,12))
sns.barplot(x='month', y = 'daily_non_vehicle_traffic', data = visual_data)
plt.show()
```

```python
#total_vehicle_volume_daily plotted against weekday. 
plt.figure(figsize=(12,10))
sns.boxplot(x='day_of_week', y='total_vehicle_volume_daily', data = visual_data)
plt.show()
```

```python
#total_vehicle_volume_daily plotted against weekday. 
plt.figure(figsize=(12,10))
sns.barplot(x='day_of_week', y='total_vehicle_volume_daily', data = visual_data)
plt.show()
```

```python
#daily_non_vehicle_traffic plotted against weekday. 
plt.figure(figsize=(12,10))
sns.boxplot(x='day_of_week', y='daily_non_vehicle_traffic', data = visual_data)
plt.show()
```

```python
#daily_non_vehicle_traffic plotted against weekday. 
plt.figure(figsize=(12,10))
sns.barplot(x='day_of_week', y='daily_non_vehicle_traffic', data = visual_data)
plt.show()
```

```python
#Exploring avg_temp_f_daily on month
plt.figure(figsize=(12,10))
sns.barplot(y='avg_temp_f_daily',x='month', data = visual_data)
plt.show()
```

```python
#aggreagating total_vehicle_volume_daily over month and plotting 

visual_data_line = visual_data.groupby('month').aggregate({'total_vehicle_volume_daily':'mean'})
plt.figure(figsize=(12,10))
sns.lineplot(x = visual_data_line.index, y = visual_data_line.total_vehicle_volume_daily, data = visual_data_line)
plt.show()
```

```python
#aggreagating daily_non_vehicle_traffic over month and plotting 

visual_data_line2 = visual_data.groupby('month').aggregate({'daily_non_vehicle_traffic':'mean'})
plt.figure(figsize=(12,10))
sns.lineplot(x = visual_data_line2.index, y = visual_data_line2.daily_non_vehicle_traffic, data = visual_data_line2)
plt.show()
```
