```python
#Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing dataset from csv to data frame
visual_data = pd.read_csv('machine_learning_set.csv')
```

```python
visual_data
```

```python
visual_round= visual_data.round(decimals=1)
```

```python
visual_round
```

```python
#Exploring daily_non_vehicle_traffic on holidays
holidays = visual_round.loc[visual_data.holiday != 'none']
plt.figure(figsize=(10,8))
sns.boxplot(y='holiday',x='daily_non_vehicle_traffic', data = holidays)
plt.show()
```

```python
#Exploring daily_non_vehicle_traffic on holidays
holidays = visual_round.loc[visual_data.holiday != 'none']
plt.figure(figsize=(10,8))
sns.barplot(y='holiday',x='daily_non_vehicle_traffic', data = holidays)
plt.show()
```

```python
#Exploring total_vehicle_volume_daily on holidays
holidays = visual_round.loc[visual_data.holiday != 'none']
plt.figure(figsize=(10,8))
sns.boxplot(y='holiday',x='total_vehicle_volume_daily', data = holidays)
plt.show()
```

```python
#Exploring total_vehicle_volume_daily on holidays
holidays = visual_round.loc[visual_data.holiday != 'none']
plt.figure(figsize=(10,8))
sns.barplot(y='holiday',x='total_vehicle_volume_daily', data = holidays)
plt.show()
```

```python
#Plotting total_vehicle_volume_daily over month
plt.figure(figsize=(18,12))
g=sns.barplot(x='month', y = 'total_vehicle_volume_daily', data = visual_round)
plt.title("Traffic Volume by Month")
g.set_xticklabels(['January','February','March','April','May','June','July','August','September','October','November','December'])
# This will add label to X-axis
g.set_xlabel( "Month of Year") 
# This will add label to Y-axis
g.set_ylabel( "Monthly Traffic Volume")
plt.show()
```

```python
#Plotting total_vehicle_volume_daily over total_rain_mm_daily
plt.figure(figsize=(18,12))
sns.barplot(x='total_rain_mm_daily', y = 'total_vehicle_volume_daily', data = visual_round)
plt.show()
```

```python
#Plotting total_vehicle_volume_daily over total_snow_mm_daily
plt.figure(figsize=(18,12))
sns.barplot(x='total_snow_mm_daily', y = 'total_vehicle_volume_daily', data = visual_round)
plt.show()
```

```python
#Plotting daily_non_vehicle_traffic over month
plt.figure(figsize=(18,12))
g=sns.barplot(x='month', y = 'total_vehicle_volume_daily', data = visual_round)
plt.title("Non-Vehicle Volume by Month")
g.set_xticklabels(['January','February','March','April','May','June','July','August','September','October','November','December'])
# This will add label to X-axis
g.set_xlabel( "Month of Year") 
# This will add label to Y-axis
g.set_ylabel( "Monthly Non-Vehicle Traffic Volume")
plt.show()
```

```python
#total_vehicle_volume_daily plotted against weekday. 
plt.figure(figsize=(12,10))
sns.boxplot(x='day_of_week', y='total_vehicle_volume_daily', data = visual_round)
plt.show()
```

```python
#total_vehicle_volume_daily plotted against weekday. 
plt.figure(figsize=(12,10))
g=sns.barplot(x='day_of_week', y='total_vehicle_volume_daily', data = visual_round)
plt.title("Traffic Volume by Days of the Week")
g.set_xticklabels(['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'])
# This will add label to X-axis
g.set_xlabel( "Day of Week") 
# This will add label to Y-axis
g.set_ylabel( "Daily Traffic Volume")
plt.show()
```

```python
#daily_non_vehicle_traffic plotted against weekday. 
plt.figure(figsize=(12,10))
sns.boxplot(x='day_of_week', y='daily_non_vehicle_traffic', data = visual_round)
plt.show()
```

```python
#daily_non_vehicle_traffic plotted against weekday. 
plt.figure(figsize=(12,10))
g=sns.barplot(x='day_of_week', y='daily_non_vehicle_traffic', data = visual_round)
plt.title("Non-Vehicle Traffic by Days of the Week")
g.set_xticklabels(['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'])
# This will add label to X-axis
g.set_xlabel( "Day of Week") 
# This will add label to Y-axis
g.set_ylabel( "Daily Non-Vehicle Volume")
plt.show()
```

```python
#Exploring avg_temp_f_daily on month
plt.figure(figsize=(12,10))
sns.barplot(y='avg_temp_f_daily',x='month', data = visual_round)
plt.show()
```

```python
#aggreagating total_vehicle_volume_daily over month and plotting 

visual_data_line = visual_round.groupby('month').aggregate({'total_vehicle_volume_daily':'mean'})
plt.figure(figsize=(12,10))
sns.lineplot(x = visual_data_line.index, y = visual_data_line.total_vehicle_volume_daily, data = visual_data_line)
plt.show()
```

```python
#aggreagating daily_non_vehicle_traffic over month and plotting 

visual_data_line2 = visual_round.groupby('month').aggregate({'daily_non_vehicle_traffic':'mean'})
plt.figure(figsize=(12,10))
sns.lineplot(x = visual_data_line2.index, y = visual_data_line2.daily_non_vehicle_traffic, data = visual_data_line2)
plt.show()
```

```python
sns.regplot(x="total_vehicle_volume_daily", y="avg_temp_f_daily", data=visual_round)
```

```python
sns.lmplot(x="total_vehicle_volume_daily", y="avg_temp_f_daily", col="day_of_week", data=visual_round,
           aspect=1);
```

```python
sns.lmplot(x="daily_non_vehicle_traffic", y="avg_temp_f_daily", col="day_of_week", data=visual_round,
           aspect=1);
```

```python
sns.catplot(x="day_of_week", y="daily_non_vehicle_traffic", kind="swarm", data=visual_round)
```

```python
sns.catplot(x="day_of_week", y="total_vehicle_volume_daily", kind="swarm", data=visual_round)
```
